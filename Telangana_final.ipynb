```python
def find_patterns(variable,cell):
    found_patterns=[]
    for pattern in variable:
        match=re.search(pattern,cell)
        if match:
            found_patterns.append(match.group(0))
            break
        else:
            pass
    final=''.join(found_patterns)
    return final
```
