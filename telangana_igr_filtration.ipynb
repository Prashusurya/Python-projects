```python
def extract_buyers_sellers(text):
    entries = re.split(r'(\d+\.)', text)
    entries = [entry.strip() for entry in entries if entry.strip() and not re.match(r'\d+\.', entry)]
    buyers = ', '.join(entry.replace('(Buyer)', '').strip() for entry in entries if '(Buyer)' in entry)
    sellers = ', '.join(entry.replace('(Seller)', '').strip() for entry in entries if '(Seller)' in entry)
    return pd.Series([buyers, sellers])
```
