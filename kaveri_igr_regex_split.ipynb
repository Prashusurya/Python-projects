```python
def generate_N_grams(text,ngram):
    words=[word for word in text.split(" ")]
    temp=zip(*[words[i:] for i in range(0,ngram)])
    ans=[' '.join(ngram) for ngram in temp]
    return ans
```

```python
def filter(text):
    res = re.sub(' +', ' ', text)
    res = re.sub(',', ' ', res)
    res = re.sub("'", '', res)
    res = re.sub('"', '', res)
    res = re.sub('sq. ft.|sq. ft|sq.ft.|sq ft|sq. feet|square feet|sq feet|square ft|sqft.|sq.feet|sq fet|sq.ft|sq fnt', ' sqft ', res)
    res = re.sub('sq mtrs|sq. Mtrs.|sqi mtrs|Sq. meters|sq.mtr|sqmtr|sqmts|square mtr|sqm', ' sqmt ', res)
    res = re.sub('meters|mtrs', ' mtr ', res)
    res = re.sub('super built-up area|super built up area|super built-up|super built up|super builtup|super built area|super b', ' sba  ', res)
    res = re.sub('built-up area|build up area|built-up|built up|builtup', ' sba  ', res)
    res = re.sub('built area', ' sba  ', res)
    res = re.sub('carpet area', ' carpet ', res)
    res = re.sub('carpetarea', ' carpet ', res)
    res = re.sub('undivided share', ' uds ', res)
    res = re.sub(' +', ' ', res)
    return res
```
