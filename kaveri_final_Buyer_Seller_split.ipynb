```python
def company_name(x):
    company=[]
    if len(x)==1:
        return x
    elif len(x)>1:
        for j in range(0,len(x)):
            if any(word in x[j] for word in company_keywords):
                return x[j]
                break
            else:
                pass
        return x[0]
    else:
        return None
    return company
```

```python
def single_company(x):
    output = ''
    words = x.split()
    matched = False
    for word in words:
        normalized_word = word.lower().strip('.,()')
        if normalized_word in company_keywords:
            matched = True
        elif matched:
            break
        output += ' ' + word if output else word
    return output
```
